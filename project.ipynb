```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the CSV
df = pd.read_csv('data.csv')

# Show the first few rows to make sure it worked
df.head()
```

```python
# X is the input (Features), y is what we want to predict (Target)
X = df[['Hours_Studied']] 
y = df['Final_Grade']

# Create the Model
model = LinearRegression()

# Train the model (The "Learning" phase)
model.fit(X, y)

print("Model trained successfully!")
```

```python
# Predict for 6 hours of study
hours = [[6]]
predicted_grade = model.predict(hours)

print(f"If you study for 6 hours, your predicted grade is: {predicted_grade[0]:.2f}")

# Bonus: Visualize it!
plt.scatter(X, y, color='blue') # Real data points
plt.plot(X, model.predict(X), color='red') # The "Line of Best Fit"
plt.xlabel('Hours Studied')
plt.ylabel('Final Grade')
plt.title('Study Hours vs Grade')
plt.show()
```

```python
# The same visualization but looking cleaner
plt.figure(figsize=(8, 5)) # Makes the graph a bit bigger
plt.scatter(X, y, color='blue', label='Actual Student Data') 
plt.plot(X, model.predict(X), color='red', linewidth=2, label='ML Prediction Line')

# Adding labels so it looks professional
plt.xlabel('Hours Studied')
plt.ylabel('Final Grade')
plt.title('Study Hours vs Grade: Linear Regression Model')
plt.legend() # This shows the labels for the blue dots and red line
plt.grid(True, linestyle='--', alpha=0.6) # Adds a light grid
plt.show()
```
